# Data Cleaning

#### We have to merge the confirmed ,death and recovery for better  Explotary data analysis

```python
# Create new date columns
Dates = confirmed.columns[4:]
Dates
```

```python
confirmed_long = confirmed.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=Dates,
    var_name='Date',
    value_name='Confirmed')
confirmed_long
```

```python
deaths_long = death.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=Dates,
    var_name='Date',
    value_name='Deaths')
deaths_long
```

```python
recovered_long = recovery.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    value_vars=Dates,
    var_name='Date',
    value_name='Recovered')
recovered_long
```

```python
# Merging all the three dataframe  into make new data frame
data = confirmed_long.merge(
  right=deaths_long,
  how='left',
  on=['Province/State', 'Country/Region', 'Date', 'Lat', 'Long']
)
```

```python
data = data.merge(
  right=recovered_long,
  how='left',
  on=['Province/State', 'Country/Region', 'Date', 'Lat', 'Long']
)
```

# Time Series

```python
# filter the 3 dataframe for data of US only
US_confirmed=confirmed[confirmed['Country/Region']=="US"].T
US_deaths=death[death['Country/Region']=="US"].T
US_recovered=recovery[recovery['Country/Region']=="US"].T
print(US_confirmed)
print(US_deaths)
print(US_recovered)
```

```python
US_join=US_confirmed.join(US_deaths,how='left',lsuffix='_confirmed',rsuffix='_deaths')
US=US_join.join(US_recovered,how='left',lsuffix='_',rsuffix='_recovered')
US=US.rename(columns={"242_confirmed":"confirmed","242_deaths":"deaths",229:"recovered"})
US
```

```python
US=US[4:]
```

```python
US
```

```python
US.index=pd.to_datetime(US.index)
US.head()
```

## Split the train and test data

```python
# Creating train and test data
split_point=int(0.75*len(US))
print(split_point)
train=US[0:split_point]
test=US[split_point:]
```

```python
#visualize confirmed & death case for both train & test data
plt.figure(figsize=(7,5))
plt.plot(train.confirmed,label='Train confirmed case')
plt.plot(test.confirmed,label='Test confirmed case')
plt.plot(train.deaths,label='Train death')
plt.plot(test.deaths,label='Test death')
plt.tight_layout()
plt.legend()
```

## Model Creations

### Average method

```python
# Average method
pred=train.confirmed.mean()
pred=pred*len(test)
test['Average_pred']=pred
```

```python
from sklearn import metrics
from math import sqrt
mse=metrics.mean_squared_error(test.confirmed,test.Average_pred)
rmse=np.sqrt(mse)
mae=metrics.mean_absolute_error(test.confirmed,test.Average_pred)
r2=metrics.r2_score(test.confirmed,test.Average_pred)
print("R2 score",r2)
print("Mean square error",mse)
print("Root mean square error",rmse)
print("Mean absolute error",mae)
```

```python
plt.plot(test.confirmed,label="test data")
plt.plot(test.Average_pred,label="predict data")
plt.xticks(rotation=90)
plt.legend()
```

```python
# walk forward validation
X=train.confirmed.values

predict=[]
for i in test.confirmed:
    pred=X.mean()
    predict.append(pred)
    X=np.append(X,i)
```

```python
# Assign the predictions to the 'predict_Avg_Wf' column in the 'test' DataFrame
test['predict_Avg_Wf'] = predict
```

```python
mse=metrics.mean_squared_error(test.confirmed,test.predict_Avg_Wf)
rmse=np.sqrt(mse)
mae=metrics.mean_absolute_error(test.confirmed,test.predict_Avg_Wf)
r2=metrics.r2_score(test.confirmed,test.predict_Avg_Wf)
print("r2 score",r2)
print("Mean square error",mse)
print("Root mean square error",rmse)
print("Mean absolute error",mae)
```

```python
plt.plot(test.confirmed,label="test data")
plt.plot(test.predict_Avg_Wf,label="predict data")
plt.xticks(rotation=90)
plt.legend()
```

## Insights

    * The average method's performance on COVID data is poor, as indicated by an R-squared (R2) score of -23.52, a high Mean 
    Square Error (MSE) of 15,703,387,081,967, and a large Root Mean Square Error (RMSE) of approximately 3,962,750. 
    * These metrics signify substantial prediction errors, and a visual inspection of the plot shows that the test and 
    prediction values closely resemble parallel lines, further confirming the model's inadequacy.

# ARIMA

```python
#Geeting the optimal values of p,q an d
import itertools
p =d= q=range(0,5)#values of p,d,q range from 0 to 4
pdq = list(itertools.product(p,d,q))
pdq
```

```python
from statsmodels.tsa.arima.model import ARIMA
list1=[]
list2=[]
for params in pdq: #iterating params over pdq
    try:
        model_arima = ARIMA(train.confirmed, order=params)#training model on various pdq values
        model_arima_fit = model_arima.fit()#training
        print(params, model_arima_fit.aic)#printing parameter and aic values
        list1.append(params)
        list2.append(model_arima_fit.aic)
    except:
        continue
```

```python
min(list2)
```

```python
loc=list2.index(min(list2))
list1[loc]
```

```python
## create the final model with lowest aic score parameter
model_arima = ARIMA(train.confirmed, order=(3, 0, 0))
model_arima_fit = model_arima.fit() #training
```

```python
# forecast
forecast=model_arima_fit.predict(start=183,end=243)
```

```python
# visualization for forecast & test
plt.plot(forecast,label='forecast')
plt.plot(test.confirmed,label='test')
plt.xticks(rotation=90)
plt.legend()
```

```python
# walk forward validation
pred_ARIMA=[]
X=train.confirmed.values

for i in test.confirmed:
    model=SARIMAX(X,order=(1,1,1))
    model_fit=model.fit()
    pred=model_fit.forecast()
    pred_ARIMA.append(pred)
    X=np.append(X,i)
```

```python
test["pred_ARIMA"]=pred_ARIMA
```

```python
mse=mean_squared_error(test.confirmed,pred_ARIMA)
print("mean square error",mse)
rmse=np.sqrt(mse)
print("Root mean square error",rmse)
mae=mean_absolute_error(test.confirmed,pred_ARIMA)
print("Mean absolute error",mae)
r2=r2_score(test.confirmed,pred_ARIMA)
print("r2 score",r2)
```
